### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = loja['Preço'].sum()
print(f'Loja: {faturamento_loja:,.2f}')

faturamento_loja2 = loja2['Preço'].sum()
print(f'Loja2: {faturamento_loja2:,.2f}')

faturamento_loja3 = loja3['Preço'].sum()
print(f'Loja3: {faturamento_loja3:,.2f}')

faturamento_loja4 = loja4['Preço'].sum()
print(f'Loja4: {faturamento_loja4:,.2f}')
```

# 2. Vendas por Categoria


```python
vendas_por_categoria_loja = loja['Categoria do Produto'].value_counts()
print(f'Loja\n{vendas_por_categoria_loja} \n')

vendas_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
print(f'Loja2\n{vendas_por_categoria_loja2} \n')

vendas_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
print(f'Loja3\n{vendas_por_categoria_loja3} \n')

vendas_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()
print(f'Loja4\n{vendas_por_categoria_loja4} \n')
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacoes_loja = loja['Avaliação da compra'].mean()
print(f'Loja: {media_avaliacoes_loja:,.2f}')

media_avaliacoes_loja2 = loja2['Avaliação da compra'].mean()
print(f'Loja2: {media_avaliacoes_loja2:,.2f}')

media_avaliacoes_loja3 = loja3['Avaliação da compra'].mean()
print(f'Loja3: {media_avaliacoes_loja3:,.2f}')

media_avaliacoes_loja4 = loja4['Avaliação da compra'].mean()
print(f'Loja4: {media_avaliacoes_loja4:,.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_loja = loja['Produto'].value_counts()
menos_vendido_loja = vendas_loja.idxmax()
mais_vendido_loja = vendas_loja.idxmin()
print(f'Loja\nMais vendido: {mais_vendido_loja}\nMenos vendido: {menos_vendido_loja} \n')

vendas_loja2 = loja2['Produto'].value_counts()
menos_vendido_loja2 = vendas_loja2.idxmax()
mais_vendido_loja2 = vendas_loja2.idxmin()
print(f'Loja2\nMais vendido: {mais_vendido_loja2}\nMenos vendido: {menos_vendido_loja2} \n')

vendas_loja3 = loja3['Produto'].value_counts()
menos_vendido_loja3 = vendas_loja3.idxmax()
mais_vendido_loja3 = vendas_loja3.idxmin()
print(f'Loja3\nMais vendido: {mais_vendido_loja3}\nMenos vendido: {menos_vendido_loja3} \n')

vendas_loja4 = loja4['Produto'].value_counts()
menos_vendido_loja4 = vendas_loja4.idxmax()
mais_vendido_loja4 = vendas_loja4.idxmin()
print(f'Loja4\nMais vendido: {mais_vendido_loja4}\nMenos vendido: {menos_vendido_loja4} \n')
```

# 5. Frete Médio por Loja


```python
media_frete_loja = loja['Frete'].mean()
print(f'Loja: {media_frete_loja:,.2f}')

media_frete_loja2 = loja2['Frete'].mean()
print(f'Loja2: {media_frete_loja2:,.2f}')

media_frete_loja3 = loja3['Frete'].mean()
print(f'Loja3: {media_frete_loja3:,.2f}')

media_frete_loja4 = loja4['Frete'].mean()
print(f'Loja4: {media_frete_loja4:,.2f}')
```
